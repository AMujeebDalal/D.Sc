```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
x = np.arange(1,11)
x
```

```python
y1 = 2*x
y2 = 3*x
```

```python
#color,width... optional
plt.plot(x,y1,color='orange',linewidth=2,linestyle=':')
plt.plot(x,y2,color='green',linewidth=2,linestyle=':')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
#bar-plot
student = {'Sam':30, 'Bob':50, 'Julia':90}
names = list(student.keys())
marks = list(student.values())
```

```python
names,marks
```

```python
plt.bar(names,marks,color='purple')
plt.title("Marks of Students")
plt.xlabel("Names of Students")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,marks,color='yellow')
plt.title("Marks of Students")
plt.xlabel("Names of Students")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#scatter-plot
g = [4,7,3,8,9,1,6]
h1 = [1,7,8,3,7,5,2]
h2 = [3,6,7,3,5,9,4]
plt.grid(True)
plt.scatter(g,h1,color='red')
plt.scatter(g,h2)
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
plt.hist(iris['sepal.length'],bins=20)
plt.show()
```

```python
#bar-graph is used to understand a categorical variable.
#histogram from continuous variable.
```

```python
plt.hist(iris['petal.width'],bins=20,color='red')
plt.show()
```

```python
#Box-plots:how a continous variable changes with respect to a categorical value
iris.boxplot(column='sepal.length', by='variety')
```

```python
#In graph above middle line = median line upper is 75percentile lower is 25.
```

```python
iris.boxplot(column='petal.width', by='variety')
```

```python
#using seaborn
sb.boxplot(x=iris['variety'],y=iris['sepal.length'])
```

```python
#using pichart
fruits = ['apple', 'mango', 'orange', 'bananas']
cost = [60,40,80,30]
```

```python
#utopct, shadow...optional
plt.figure(figsize=(6,6)) #changes size
plt.pie(cost,labels=fruits,autopct='%0.2f%%',shadow=True)
plt.show()
```

```python
Students = ['Mabrook', 'Afaaf', 'Onaisa']
cost = [60,40,80]
plt.figure(figsize=(4,4)) #changes size
plt.pie(cost,labels=Students,autopct='%0.2f%%',shadow=True)
plt.show()
```
